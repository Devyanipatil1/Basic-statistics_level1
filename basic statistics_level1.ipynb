**Q22**

```python
# z score for 90% CI
z1=stats.norm.ppf(0.95)
print('Z score of 90% confident interval = ',z1)

# z score for 94% CI
z2=stats.norm.ppf(0.97)
print('Z score of 94% confident interval = ',z2)

# z score for 60% CI
z3=stats.norm.ppf(0.70)
print('Z score of 60% confident interval = ',z3)
```

**Q23**

```python
#t-score for 95% CI for sample size of 25
t1=stats.t.ppf(0.975,24)
print('T score of 95% confident interval = ',t1)

#t-score for 96% CI for sample size of 25
t2=stats.t.ppf(0.98,24)
print('T score of 96% confident interval = ',t2)

#t-score for 99% CI for sample size of 25
t3=stats.t.ppf(0.995,24)
print('T score of 99% confident interval = ',t3)
```

**Q24**

```python
#H0: Average life of bulb >= 270
#Ha: Average life of bulb <270

#Mean=270
#s_mean=260
#std=90
#n=18

t=(260-270)/(90/np.sqrt(18))

pvalue=stats.t.cdf(np.round(t,3),df=17)

print('The probabilty of getting an average life of no more than 260 days are',np.round((pvalue)*100,3))

```
